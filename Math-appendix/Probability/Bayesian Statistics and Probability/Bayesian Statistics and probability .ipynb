# What are Frequentist Statistics
**Frequentist Statistics** test whether an event (hypothesis) occurs or not. It calculates the probability of an event in the long run of the experiment (i.e the experiment is repeated under the same conditions to obtain the outcome).

Here, the sampling distributions of **fixed size** are taken. Then, the experiment is theoretically repeated **infinite number of times** but practically done with a stopping intention. For example, I perform an experiment with a stopping intention in mind that I will stop the experiment when it is repeated 1000 times or I see minimum 300 heads in a coin toss.

Lets look at the example of a coin toss. We are trying to estimate the fairness of the coin. 



* We know that probability of getting a head on tossing a fair coin is 0.5. 
* No. of heads represents the actual number of heads obtained. 
* Difference is the difference between 0.5*(No. of tosses) - no. of heads.

An important thing is to note that, though the difference between the actual number of heads and expected number of heads( 50% of number of tosses) increases as the number of tosses are increased, the proportion of number of heads to total number of tosses approaches 0.5 (for a fair coin).

This experiment presents us with a very common flaw found in frequentist approach i.e. **Dependence of the result of an experiment on the number of times the experiment is repeated.**

# Flaws in Frequentist Statistics
### 1) p-values depend on sample size
The p-value is the probability of obtaining an effect at least as extreme as that in your data, *if* the null hypothesis is true. Well, p-values measured against a sample (fixed size) statistic with some stopping intention change with change in intention and sample size. i.e If two people work on the same data and have a different stopping intention, they may get two different  p- values for the same data, which is undesirable.

For example: Person A may choose to stop tossing a coin when the total count reaches 100 while B stops at 1000. For different sample sizes, we get different t-scores and different p-values. Similarly, intention to stop may change from fixed number of flips to total duration of flipping. In this case too, we are bound to get different p-values.

### 2) Confidence Interval (C.I) depends on sample size
Like p-value, the confidence interval depends heavily on the sample size. This makes the stopping potential absolutely absurd since no matter how many people perform the tests on the same data, the results should be consistent.

### 3) Confidence Intervals are not probability distributions
Therefore they do not provide the most probable value for a parameter and the most probable values. They simply say that there is a percentage chance that the real value lies within that interval.

# Bayes Theorem Example
Lest say we are faced with the following test scenario:
* 1% of women have breast cancer (and therefore 99% do not).
* 80% of mammograms detect breast cancer when it is there (and therefore 20% miss it).
* 9.6% of mammograms detect breast cancer when it’s not there (and therefore 90.4% correctly return a negative result).

We can put this into a table, that looks like this: 



Here is how we read the table: 
* 1% of women have cancer
* If you already have cancer, you are in the first column. There’s an 80% chance you will test positive. There’s a 20% chance you will test negative.
* If you don’t have cancer, you are in the second column. There’s a 9.6% chance you will test positive, and a 90.4% chance you will test negative.

How could we represent this graphically...



Keep that diagram in mind as we go through this example, it is invaluable in thinking of the situation intuitively. 

### How accurate is the test? 
Now suppose you get a positive test result. What are the chances you have cancer? 80%? 99%? 1%?

Lets walk through it. 
* Ok, we got a positive result. It means we’re somewhere in the top row of our table. Let’s not assume anything — it could be a true positive or a false positive.
* The chances of a true positive = chance you have cancer * chance test caught it = 1% * 80% = .008
* The chances of a false positive = chance you don’t have cancer * chance test caught it anyway = 99% * 9.6% = 0.09504
Lets update our table with this information:



What was our question again? **What’s the chance we really have cancer if we get a positive result?**. Well, the chance of an event is the number of ways it could happen given all possible outcomes. This leads us to some general intuition about probability! 

### $$Probability = \frac{desired \; event}{all \; possibilities}$$

What does that look like in our case? Well we have:
* desired event = given a positive test, we have cancer
* all possibilities = a postive test and having cancer + a positive test and having no cancer

In other words:

### $$Probability \; of \;cancer \;given\;positive\;test = \frac{positive\;test\;and\;cancer}{positive\;test\;and\;cancer+positive\;test\;and\;no\;cancer}$$

And if we fill in our values from the table we arrive at:

### $$Probability = \frac{0.01*0.8}{0.01*0.8+0.99*0.096}=0.0776=7.6\%$$

**Interesting!** — a positive mammogram only means you have a 7.8% chance of cancer, rather than 80% (the supposed accuracy of the test). It might seem strange at first but it makes sense: the test gives a false positive 9.6% of the time (quite high), so there will be many false positives in a given population. For a rare disease, most of the positive test results will be wrong.

Let’s test our intuition by drawing a conclusion from simply eyeballing the table. If you take 100 people, only 1 person will have cancer (1%), and they’re most likely going to test positive (80% chance). Of the 99 remaining people, about 10% will test positive, so we’ll get roughly 10 false positives. Considering all the positive tests, just 1 in 11 is correct, so there’s a 1/11 chance of having cancer given a positive test. The real number is 7.8% (closer to 1/13, computed above), but we found a reasonable estimate without a calculator.

# Probability Rules 
Lets look at a few probability rules and try to come back and solve the above example.
## Conditional Probability
This defines the probabilty of event A occuring, given that event B occurs. In this case A and B are dependent. Mathematically it looks like: 
### $$P(A|B) = \frac{P(A\cap B)}{P(B)}$$
and we can use the multiplication rule:
### $$P(A \cap B) = P(A)*P(B|A)$$
to get:
### $$P(A|B)=\frac{P(A)*P(B|A)}{P(B)}$$
This is known as **Bayes Theorem!!!**

An example: Given that you draw a red card, what is the probability that it is a 4?
### $$P(4|red) = \frac{P(4)*P(red|4)}{P(red)}=\frac{\frac{1}{13}*\frac{1}{2}}{\frac{1}{2}}=\frac{1}{13}$$

## Joint Probability
This defines the probability of event A occuring and event B occuring, otherwise defined as the intersection of A and B. Mathematically it looks like:
### $$P(A\;and\;B)=P(A \cap B)$$
Example: If you draw a card at random, what is the probability that it is 4 and red?
### $$P(4\cap red) = \frac{4}{52}*\frac{26}{52} = \frac{1}{26}$$

## Conditional vs joint
The main thing that changes between conditional and joint probability is that when we work with conditional the sample space changes because you have been given new information, so some choices can be eliminated and certain probabilities may update! 

# Back to our example, using Bayes Theorem
Okay, armed with our new knowledge, lets try and solve our orignal problem and see if we end up with the original answer!

Remember, Bayes theorem is defined as: 
### $$P(A|B)=\frac{P(A)*P(B|A)}{P(B)}$$
In our example that would look like:
### $$P(cancer\;|\;positive\;test)=\frac{P(cancer)*P(positive\;test\;|\;cancer)}{P(positive\;test)}$$
### $$\frac{0.01*0.8}{0.01*0.8+0.99*0.096}=7.8\%$$
This yields the same result as our general intuition equation! Maybe Bayes theorem isn't so hard after all...

# Bayesian Inference Example
Lets look at the example of weighing a dog at the vet's office. When we take the dog to the vet, it squirms on the scale which makes it difficult to get an accurate reading. 

At the last visit, we got 3 readings for its weight:
* 13.9 lb
* 17.5 lb
* 14.1 lb

There is a standard statistical interpretation for this. We are able to calculate the mean, standard deviation, and standard error for this set of numbers and create a distribution for the dogs actual weight!

### Find the mean

```python
mu = (13.9+17.5+14.1)/3
```

### Find the Standard Deviation (sample)

```python
import numpy as np
sigma = np.sqrt( (np.power((13.9-mu),2)+np.power((17.5-mu),2)+np.power((14.1-mu),2)) /(3-1)) 
```

### Find the Standard Error

```python
std_error = sigma / np.sqrt(3)
```

### Normal Distribution Background
At this point, we have found the metrics for the data set that we need in order to find determine the normal distribution. The probability density function for a normal distribution looks like: 



### Central Limit Theorem
The normal distribution is incredibly useful due to the **Central Limit Theorem**, which states that the **sampling distribution** of the **sampling means** approaches a **normal distribution** as the sample size gets **larger** — *no matter what the shape of the population distribution*. This fact holds especially true for sample sizes over 30. All this is saying is that as you take more samples, especially large ones, your graph of the sample means will look more like a normal distribution.

“Many real-world observations can be approximated by, and tested against, the same expected pattern: the normal distribution. In this familiar symmetric bell-shaped pattern, most observations are close to average, and there are fewer observations further from the average. The size of flowers, the physiological response to a drug, the breaking force in a batch of steel cables — these and other observations often fit a normal distribution.

There are, however, many important things we would like to measure and test that do not follow a normal distribution. Household income doesn’t – high values are much further from the average than low values are.

But even when raw data does not fit a normal distribution, there is often a normal distribution lurking within it. This makes it possible to still use the normal distribution to test ideas about non-normal data.”

### Central Limit Theorem Defined
Some data sets follow a normal distribution. Others do not. However for both normal AND nonnormal data, if we repeatedly take independent random samples of size n from a population, then when n is large, the distribution of the **sample means** will approach a normal distribution!

How large is large enough? Well, it depends. The closer the population distribution is to a normal distribution, the fewer samples you need to take to demonstrate the theorem. Populations that are heavily skewed or have several modes may require larger sample sizes.

```python
#imports
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import math
import seaborn as sns
%matplotlib inline 

# styling
# Seaborn Plot Styling
sns.set(style="white", palette="husl")
sns.set_context("poster")
sns.set_style("ticks")
fig, ax = plt.subplots(figsize=(12,8))

# plotting distribution
x = np.linspace(mu - 3.5*std_error, mu + 3.5*std_error, 100)
plt.plot(x,mlab.normpdf(x, mu, std_error))

# plotting mean
plt.axvline(x=mu, color="blue",  ymax=(mlab.normpdf(mu, mu, std_error))/.36)

# plotting standard error
plt.axvline(x=mu-std_error, color="blue", ymax=(mlab.normpdf((mu-std_error), mu, std_error))/.35)
plt.axvline(x=mu+std_error, color="blue", ymax=(mlab.normpdf((mu+std_error), mu, std_error))/.35)

# plotting our measurements
plt.axvline(x=13.9, color="green", ymax=(mlab.normpdf(13.9, mu, std_error))/.35)
plt.axvline(x=17.5, color="green", ymax=(mlab.normpdf(17.5, mu, std_error))/.29)
plt.axvline(x=14.1, color="green", ymax=(mlab.normpdf(14.1, mu, std_error))/.35)

# setting axes
ax.set_xlabel('Weight in Pounds', fontsize=20)
ax.set_ylabel('Probability', fontsize=20)
ax.tick_params(labelsize="large")
ax.set_title('Normal Distribution of Dogs weight',fontsize=20)
ax.legend(['Weight Distribution', 'mean', 'standard error', 'standard error', 'measurements'], fontsize=20)

plt.show()
```

# Bayes Theorem
Bayes Theorem is very useful in making the most out of small data sets. Lets take a minute to go over what it looks like again. 
### $$P(A|B)=\frac{P(B|A)*P(A)}{P(B)}$$
or in our case: 
### $$P(weight\;|\;measurements)=\frac{P(measurements\;|\;weight)*P(weight)}{P(measurements)}$$
simplified to:
### $$P(w\;|\;m)=\frac{P(m\;|\;w)*P(w)}{P(m)}$$
We can clearly see 4 terms above. Each of these terms represents a different part of the process.

---

### P(w) - Prior
* this represents our **prior beliefs**
* In this case it represents what we believe about the dogs weight before we even put it on the scale
* "probability of cancer before you know the test result"

### P(m | w) - Likelihood
* This is the **likelihood** that our measurements would occur, given a particular weight
* example: given that the dogs weight is 200lbs, what is the likelihood that we measure her to be 13.9,14.1, and 17.5 lbs?
* "if you know the patient has cancer, how likely are you to get a positive test result?"

### P(w | m) - Posterior
* This is the **posterior**, which shows the probability of the dog being a weight, given the measurements that we made 
* this is what we are most interested in finding 
* "What do we believe the probability of cancer is, given the test result?"

### P(m)
* this is the probability of data, showing the probability that any given data point will be measured
* we will assuming this is constant, i.e. that the scale is unbiased 

---

We can think of this formula as saying: "lets start with a belief $P(w)$, take some measurements and update it with $P(m|w)$, and then we have a new belief once done-$P(w|m)$. 

Okay great, so at this point, it is not a terrible idea to start by perfectly agnostic and making no assumptions about the result. In this case, we assume that the dogs weight is equally likely to be 13 lbs, or 15 lbs, or 1 lbs, or 100000 lbs, and we will let the data speak. 

To do this we assume a uniform prior, meaning that its probability is a constant for all values. This allows us to reduce our equation to:
### $$P(w\;|\;m) = P(m\;|\;w)$$

So we now want to calculate this- that is we want to calculate the probability of our measurements occuring, given a weight. We will want to do this for all of the possible weights! 

### Calculation
So this is what our equation looks like, given the weight 17lbs.
### $$P(w=17\;|\;m=[13.9,14.1,17.5] = P(m=[13.9,14.1,17.5]\;|\;w=17))$$
So, given a weight of 17 lbs, what does our probability distribution look like?

```python
# styling
fig, ax = plt.subplots(figsize=(12,8))

# setting average weight to be 17, because this is the weight that we are given 
mu_17 = 17

# Create a distribution that models our beliefs based around the dogs weight being given as 17 lbs
# We can see that the mean is 17 lbs in this case, and the weight distribution is created from the 
# pdf discussed earlier
x = np.linspace(mu_17 - 3.5*std_error, mu_17 + 3.5*std_error, 100)
plt.plot(x,mlab.normpdf(x, mu_17, std_error))

# plotting mean
plt.axvline(x=mu_17, color="blue",  ymax=(mlab.normpdf(mu_17, mu_17, std_error))/.36)

# plotting standard error
plt.axvline(x=mu_17-std_error, color="blue", ymax=(mlab.normpdf((mu_17-std_error), mu_17, std_error))/.35)
plt.axvline(x=mu_17+std_error, color="blue", ymax=(mlab.normpdf((mu_17+std_error), mu_17, std_error))/.35)

# plotting our measurements
plt.axvline(x=13.9, color="green", ymax=(mlab.normpdf(13.9, mu_17, std_error))/.15)
plt.axvline(x=17.5, color="green", ymax=(mlab.normpdf(17.5, mu_17, std_error))/.35)
plt.axvline(x=14.1, color="green", ymax=(mlab.normpdf(14.1, mu_17, std_error))/.17)

# setting axes
ax.set_xlabel('Weight in Pounds', fontsize=20)
ax.set_ylabel('Probability', fontsize=20)
ax.set_title('Weight = 17bs',fontsize=20)
ax.legend(['Weight Distribution', 'mean', 'standard error', 'standard error', 'measurements'], fontsize=20)
ax.tick_params(labelsize="large")
plt.show()
```

We can see our measurements here again in green, as well as the mean 17lbs, and a standard deviation for each. So, to calculate the probability, or *likelihood* in this case, of our measurements occuring given this weight of 17 lbs, we do the following:
### $$P(m=13.9|w=17)*P(m=14.1|w=17)*P(m=17.5|w=17)$$
Lets walk through this in code:

```python
# find the likelihood of measuring 13.9 if the weight is 17
p139 = mlab.normpdf(13.9, mu_17, std_error)

# find the likelihood of measuring 14.1 if the weight is 17
p141 = mlab.normpdf(14.1, mu_17, std_error)

# find the likelihood of measuring 17.5 if the weight is 17
p175 = mlab.normpdf(17.5, mu_17, std_error)

# total likelihood of measurements
p_total_17 = p139*p141*p175
p_total_17
```

```python
# input weights that we will plug into above formula in an attempt to optimize
input_weight = np.arange(13,19,.1)

# creating array to hold our likelihood values for each input weight
output_likelihood = np.zeros(60)

index = 0
for weight in np.nditer(input_weight):
    mu = weight
    output_likelihood[index] = mlab.normpdf(13.9, mu, std_error) * mlab.normpdf(14.1, mu, std_error) * mlab.normpdf(17.5, mu, std_error)
    index += 1

print(('What weight makes the measurements most likely: '), input_weight[np.argmax(output_likelihood)])
```

# Non-Uniform Prior - Think regularization
Lets start with a bit of background information! The last time that we came into the vet to weigh the dog, we found that it weighed **14.2 lbs**, with a standard error of **1 lb**. At this point she doesn't seem noticably more heavy than last time. 

So lets define this is as our **prior**, or $p(w)$. It takes the following form: 

```python
# styling
fig, ax = plt.subplots(figsize=(12,8))

# setting our prior weight to be 14.2 lbs, with a standard error of 1 lb
mu_prior = 14.2
std_error_prior = 1

# Create a distribution that models our beliefs based around the dogs weight being given as 14.2 lbs
x = np.linspace(mu_prior - 3.5*std_error_prior, mu_prior + 3.5*std_error_prior, 100)
plt.plot(x,mlab.normpdf(x, mu_prior, std_error_prior))

# plotting mean
plt.axvline(x=mu_prior, color="blue",  ymax=(mlab.normpdf(mu_prior, mu_prior, std_error))/.36)

# plotting standard error
plt.axvline(x=mu_prior-std_error_prior, color="blue", ymax=(mlab.normpdf((mu_prior-std_error_prior), mu_prior, std_error_prior))/.41)
plt.axvline(x=mu_prior+std_error_prior, color="blue", ymax=(mlab.normpdf((mu_prior+std_error_prior), mu_prior, std_error_prior))/.41)


# setting axes
ax.set_xlabel('Weight in Pounds', fontsize=20)
ax.set_ylabel('Probability', fontsize=20)
ax.set_title('Prior Weight = 14.2 lbs',fontsize=20)
ax.legend(['P(w) Weight Dist', 'mean', 'standard error', 'standard error', 'measurements'], fontsize=20)
ax.tick_params(labelsize="large")
plt.show()
```

This probability distribution above is what we believe about the dogs weight before we even put it on the scale. It is the prior probability of her being a given weight. Now this time around we are going to *use* the prior. 

Our equation will look like: 
### $$P(w\;|\;m)=P(m\;|\;w)*P(w)$$

The way that this plays out now is we say:
* what if she were 17 lbs? 
* we need to multiply that by probability of our prior showing that she was 17 lbs, which is quite small
* to see that, look at the prior distribution above- the probability for a weight being 17 is *very* small

What is the probability that the dog is 17 lbs, according to our prior? 

```python
# determining the probability the dog is 17 lbs according to our prior
mlab.normpdf(17, mu_prior, std_error_prior)
```

We can see that according to our prior distribution, the probability that the dog is 17 lbs is very small. 

So we need to multiply this probability times the likelihood we received out measurements, given a weight of 17 lbs. 

Mathematically that looks like: 
### $$P(w=17\;|\;m=[13.9,14.1,17.5] = P(m=[13.9,14.1,17.5]\;|\;w=17)*P(w=17))$$

Lets take a quick look at this graphically.

```python
# -------------------------------------- plot for likelihood -------------------------------------

# styling
fig, ax = plt.subplots(figsize=(15,8))

# Create a distribution that models our beliefs based around the dogs weight being given as 17 lbs
# We can see that the mean is 17 lbs in this case, and the weight distribution is created from the 
# pdf discussed earlier

x = np.linspace(mu_17 - 3.5*std_error, mu_17 + 3.5*std_error, 100)
plt.plot(x,mlab.normpdf(x, mu_17, std_error))

# plotting our measurements
plt.axvline(x=13.9, color="green", ymax=(mlab.normpdf(13.9, mu_17, std_error))/.15)
plt.axvline(x=17.5, color="green", ymax=(mlab.normpdf(17.5, mu_17, std_error))/.41)
plt.axvline(x=14.1, color="green", ymax=(mlab.normpdf(14.1, mu_17, std_error))/.19)

# -------------------------------------- plot for prior -------------------------------------

# Create a distribution that models our beliefs based around the dogs weight being given as 14.2 lbs
x = np.linspace(mu_prior - 3.5*std_error_prior, mu_prior + 3.5*std_error_prior, 100)
plt.plot(x,mlab.normpdf(x, mu_prior, std_error_prior))

# -------------------------------------- common for each plot -------------------------------------
# setting axes
ax.set_xlabel('Weight in Pounds', fontsize=20)
ax.set_ylabel('Probability', fontsize=20)
ax.set_title('Weight = 17bs',fontsize=20)
ax.legend(['Weight Dist - P(m|w)', 'measurement', 'measurement', 'measurement', 'Prior - P(w)'], fontsize=20)
ax.tick_params(labelsize="large")
plt.show()
```

Above we can see the probability distributions (beliefs) for our prior, as well as the likelihood, when the weight is 17 lbs. 

Now, lets perform the operation from the equation above, where we multiply the likelihood by the prior, in order to get our new set of beliefs for P(w|m).

```python
# -------------------------------------- plot for likelihood -------------------------------------
# styling
fig, ax = plt.subplots(figsize=(15,8))

# Create a distribution that models our beliefs based around the dogs weight being given as 17 lbs
# We can see that the mean is 17 lbs in this case, and the weight distribution is created from the 
# pdf discussed earlier
x = np.linspace(mu_17 - 3.5*std_error, mu_17 + 3.5*std_error, 100)
plt.plot(x,mlab.normpdf(x, mu_17, std_error))

# -------------------------------------- plot for likelihood, times prior -------------------------------------
# multiply likelihood times the prior probability of the dog being 17 lbs
prior_17 = mlab.normpdf(17, mu_prior, std_error_prior)
x = np.linspace(mu_17 - 3.5*std_error, mu_17 + 3.5*std_error, 100)
plt.plot(x,mlab.normpdf(x, mu_17, std_error)*prior_17, 'b')

# -------------------------------------- plot for prior -------------------------------------
# Create a distribution that models our beliefs based around the dogs weight being given as 14.2 lbs
x = np.linspace(mu_prior - 3.5*std_error_prior, mu_prior + 3.5*std_error_prior, 100)
plt.plot(x,mlab.normpdf(x, mu_prior, std_error_prior))

# -------------------------------------- common for each plot -------------------------------------
# setting axes
ax.set_xlabel('Weight in Pounds', fontsize=20)
ax.set_ylabel('Probability', fontsize=20)
ax.set_title('Weight = 17bs',fontsize=20)
ax.legend(['P(m|w)', 'P(m|w)*P(w)', 'Prior - P(w)'], fontsize=20)
ax.tick_params(labelsize="large")
plt.show()
```

Now take a look at our transformed likelihood, the blue line (P(m|w)*P(w))-its very small! 

Now if we repeat the process that we went through earlier-determining the likelihood of each measurement, given the updated likelihood distribution (blue curve), we can find the overall maximum likelihood! Lets give this a try below. 

```python
# calculate the prior probability that the weight is 17 lbs, incorporating prior
prior_17 = mlab.normpdf(17, mu_prior, std_error_prior)

# find the likelihood of measuring 13.9 if the weight is 17, incorporating prior
p139_with_prior = mlab.normpdf(13.9, mu_17, std_error)*prior_17

# find the likelihood of measuring 14.1 if the weight is 17, incorporating prior
p141_with_prior = mlab.normpdf(14.1, mu_17, std_error)*prior_17

# find the likelihood of measuring 17.5 if the weight is 17, incorporating prior
p175_with_prior = mlab.normpdf(17.5, mu_17, std_error)*prior_17

# total likelihood of measurements
p_total_17_with_prior = p139_with_prior*p141_with_prior*p175_with_prior
p_total_17_with_prior
```
