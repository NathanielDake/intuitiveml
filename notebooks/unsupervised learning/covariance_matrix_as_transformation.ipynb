* In the standard i hat j hat basis, i hat is moved to [1, 0.5] and j hat is moved to [0.5, 1]. In other words they are both stretched slightly and rotated. We can see this below.

### Pre transform, standard basis 

```python
fig, ax = plt.subplots(figsize=(11,10))

colors = ['g', 'r' , 'b']
vector_green = np.array([[1], [0]])
vector_red = np.array([[0], [1]])

ax.scatter(x,y,alpha=0.5)

vectors = [[vector_green, vector_red]]
for idx, v in enumerate(vectors[0]):
    ax.quiver([0], [0], [v[0]], [v[1]], angles='xy', scale_units='xy', scale=1, color=[colors[idx]])
```

```python
cov.dot(vector_green)
```

### Post transform, i hat and j hat are taken to...

```python
fig, ax = plt.subplots(figsize=(11,10))

colors = ['g', 'r' , 'b']
vector_green_after_cov = cov.dot(vector_green)
vector_red_after_cov = cov.dot(vector_red)

print(f'i hat after cov transform:\n {vector_green_after_cov}')
print(f'j hat after cov transform:\n {vector_red_after_cov}')

ax.scatter(res[:, 0], res[:, 1], alpha=0.5)

vectors = [[vector_green_after_cov, vector_red_after_cov]]
for idx, v in enumerate(vectors[0]):
    ax.quiver([0], [0], [v[0]], [v[1]], angles='xy', scale_units='xy', scale=1, color=[colors[idx]])
```

### Eigenbasis approach. Pre transform, basis vectors (green and red) are eigenvectors

```python
fig, ax = plt.subplots(figsize=(11,10))

colors = ['g', 'r' , 'b']
vector_green = np.linalg.eig(cov)[1][:, 0]
vector_red = np.linalg.eig(cov)[1][:, 1]

ax.scatter(x,y,alpha=0.5)

vectors = [[vector_green, vector_red]]
for idx, v in enumerate(vectors[0]):
    ax.quiver([0], [0], [v[0]], [v[1]], angles='xy', scale_units='xy', scale=1, color=[colors[idx]])
```

### Post transform via covariance matrix

```python
fig, ax = plt.subplots(figsize=(11,10))

colors = ['g', 'r' , 'b']
vector_green_after_cov = cov.dot(vector_green)
vector_red_after_cov = cov.dot(vector_red)

print(f'i hat after cov transform:\n {vector_green_after_cov}')
print(f'j hat after cov transform:\n {vector_red_after_cov}')

ax.scatter(res[:, 0], res[:, 1], alpha=0.5)

vectors = [[vector_green_after_cov, vector_red_after_cov]]
for idx, v in enumerate(vectors[0]):
    ax.quiver([0], [0], [v[0]], [v[1]], angles='xy', scale_units='xy', scale=1, color=[colors[idx]])
```

### This is the same as simply multiplying by eigen values! (see no dot product transform being applied below...)

```python
fig, ax = plt.subplots(figsize=(11,10))

colors = ['g', 'r' , 'b']
vector_green = np.linalg.eig(cov)[1][:, 0] * np.linalg.eig(cov)[0][0]
vector_red = np.linalg.eig(cov)[1][:, 1]* np.linalg.eig(cov)[0][1]

ax.scatter(res[:, 0], res[:, 1], alpha=0.5)

vectors = [[vector_green, vector_red]]
for idx, v in enumerate(vectors[0]):
    ax.quiver([0], [0], [v[0]], [v[1]], angles='xy', scale_units='xy', scale=1, color=[colors[idx]])
```

### Key Idea!
* represented in this particular basis, the covariance transformation simply stretches our green basis vector and squishes our red basis vector
* In other words, the component of our blue points that lies in the red direction will be squished via the covariance transform, while the green will be stretched
* By finding these particular red and green basis vectors (eigenbasis), we can say the following:
    * Our dataset, $X$, is the result of the covariance matrix transform being applied to our original space. 
    * This transform will always squish the red vector (one basis vector) and stretch the other (the green basis vector)
    * The transformed data, $X$, is the data that we are working with. 
    * So, we can see that the data we are working with, $X$, will always have one basis vector squished towards 0. The information/variance along that basis vector will hence be small (especially compared to the green basis vector).
    * So, by representing our data via these eigenbasis, we have transformed our points into a representation where are certain dimension can be dropped without losing much information.
* Regardless of the basis that we chose, our points are still moved the same objective distance (since distance is invariant across basis)
* The difference is that _post transform_, if we describe our points via the eigenbasis, then one of the dimensions (red basis) is squished towards 0 and we can drop without losing much information

```python
adj_mat = adj_mat.toarray()
```

```python
node_names = [f'node{x}' for x in range(adj_mat.shape[0])]
```

```python
adj_df = pd.DataFrame(data=adj_mat, index=node_names, columns=node_names)
```

```python
adj_df.head()
```

```python
eigenvalues, eigenvectors = np.linalg.eig(adj_mat)
eigenvectors = np.real(eigenvectors)
eigenvalues = np.real(eigenvalues)
```

These `eigenvectors` can be an eigenbasis, but note that they are written in the coordinate system of our original basis!

```python
sorted_eigen_idx = np.abs(eigenvalues).argsort()[::-1]

sorted_eigen_values = eigenvalues[sorted_eigen_idx]
```

```python
threshold = 0.01
np.abs(sorted_eigen_values) > threshold
```

### Change `adj_mat` to eigenbasis...
See [here](https://www.youtube.com/watch?v=P2LTAUO1TdA&t=217s). To change our basis we must:
1. Find the matrix of eigenbasis, written in the original coordinate system (we have this, it is our `eigenvectors` matrix). This matrix takes in a vector described in the eigenbasis as input, and returns a vector described wrt our original cs. This is not what we want though! We want the inverse. We want to take in a vector described in our cs and return is description in the eigenbasis cs. Again, see [here](https://youtu.be/P2LTAUO1TdA?t=530).
2. So, we need the inverse of our eigenvector basis matrix...

```python
eigenvectors_inv = np.linalg.inv(eigenvectors)
```

```python
adj_mat_in_eigen_basis = eigenvectors_inv.dot(adj_mat)
```

```python
adj_df_in_eigen_basis = pd.DataFrame(data=adj_mat_in_eigen_basis, index=node_names, columns=node_names)
```

## Super basic 4 node example

We start with a simple matrix representation and graph representation of our graph, in the standard node basis.

```python
adj_mat = np.array([
    [0, 1, 1, 1],
    [1, 0, 0, 0],
    [1, 0, 0, 1],
    [1, 0, 1, 0]
])
```

```python
g = nx.from_numpy_matrix(adj_mat)
nx.draw(g, with_labels=True)
```

```python
sns.heatmap(adj_mat)
```

Now we can calculate the eigenvectors and values.

```python
eigenvalues, eigenvectors = np.linalg.eig(adj_mat)
eigenvectors = np.real(eigenvectors)
eigenvalues = np.real(eigenvalues)
```

```python
print(eigenvectors)
```

```python
print(eigenvalues)
```

Now, we can change the basis of our adjacency matrix. This is done via the _inverse_ change of basis matrix. In this case we wish to change to the eigenbasis. Recall we define a change of basis matrix as denoting where our new basis vectors, $\vec{e}_1, \vec{e}_2, \dots$ are described in the original basis.

So, our eigenvector matrix does just that. So we take it's inverse and can use that to determine how a specific vector is described in new basis:

```python
eigenvectors.dot(adj_mat[:, 0])
```

```python
eigenvectors_inv = np.linalg.inv(eigenvectors)
```

```python
eigenvectors_inv
```

```python
eigenvectors_inv.dot(adj_mat[:, 0])
```

```python
adj_mat_in_eigenbasis = eigenvectors_inv.dot(adj_mat)
print(adj_mat_in_eigenbasis)
```

Above we still have a 4x4 matrix representing our graph. And, the first column is still node 0, the second column still node 1, and so on. The key idea is that these nodes, for instance node 0 below, are described wrt the eigenbasis!!! Let's make this clear, look at node 0: 

```python
adj_mat_in_eigenbasis[:, 0]
```

The first coefficient, -1.327 represents how much of the first eigen basis vector node 0 contains. But, and this is very important, the first eigenvector is simply represented as [1, 0, 0, 0], _not_ [-0.61162846, -0.2818452 , -0.52272073, -0.52272073]. 

```python
sns.heatmap(adj_mat_in_eigenbasis)
```

```python
for i in range(0, 4):
    print(f'Node: {i}, ', adj_mat_in_eigenbasis[:, i].std())
```
